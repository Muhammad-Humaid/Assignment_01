```python
import random

# Initializing Variables With Appropriate Data Types
Correct_Guesses = 0
Incorrect_Guesses = 0
Choice = "yes"

# While Loop To Iterate Through The Entire Body
while (Choice == "yes"):
    Loop = int(input("How Many Times Do You Want To Play This Game: "))
    
    # For Loop To Iterate Through The Set Amount Of Tries, The User Wants.
    for Counter in range (1, Loop+1):
        Value = random.randint(1,2)
        Guess = int(input("Heads Or Tails? Type '1' For Heads, Or '2' For Tails: "))
        
        # If-Else To Identify The Random Numbers
        if (Value == Guess):
            print("You Have Guessed It Correctly!")
            Correct_Guesses = Correct_Guesses + 1
        else:
            print("You Have Guessed It Incorrectly, Try Again!")
            Incorrect_Guesses = Incorrect_Guesses + 1
    
    # Choice To Validate User's Continuation    
    Choice = input("Do You Want To Continue This Game? (Yes/No): ").strip().lower()

# Printing The Final Stats
print()
print("Your Final Stats: ")
print(f"Correct Guesses: {Correct_Guesses}")
print(f"Incorrect Guesses: {Incorrect_Guesses}")
```
